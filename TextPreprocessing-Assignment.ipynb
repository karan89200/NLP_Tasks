```python
# fetching movie data through api (TMDB website)

HitLink = requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=20e4596131b0fcfa83443c2315094cc6&language=en-US&page=1")
data = HitLink.json()

```

```python

GenresUrl = requests.get("https://api.themoviedb.org/3/genre/movie/list?api_key=20e4596131b0fcfa83443c2315094cc6&language=en-US")
gene = GenresUrl.json()
gene['genres']
    
```

```python
# Define the function
def find_gen(gen_list):
    new_list = []
    for genId in gen_list:
        for jor in gene['genres']:
            if genId == jor['id']:
                new_list.append(jor['name'])
    return new_list

# Apply the function using .loc to avoid SettingWithCopyWarning
df.loc[:, 'genre_ids'] = df['genre_ids'].apply(find_gen)
```

### Text Preprocessing

#### Steps:
1. Transform to lowercase
2. Remove HTML tags
3. Remove hyperlinks or urls
4. Remove punctuations
5. Abbreviations
6. Spelling Corrections
7. Handling emojis
8. Tokenization
9. Stemming
10. Lemmatization


```python
### LowerCase
df['overview'] = df['overview'].str.lower()
df['overview'][0]
```

```python
### Remove HTML Tags ---------- regular expression

import re

def remove_htmltags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'',text)

df['overview'] = df['overview'].apply(remove_htmltags)
```

```python
df['overview'][4]
```

```python
### Remove Punctuations

import string
exclude = string.punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans('','',exclude))

# Dummy Output

df['overview'] = df['overview'].apply(remove_punctuation)
```

```python
### Fill Abbreviations

chat_words = {
   "asap": "As Soon As Possible",
    "tbd": "To Be Determined",
    "tba": "To Be Announced",
    "n/a": "Not Available",
    "imdb": "Internet Movie Database",
    "mpaa": "Motion Picture Association of America",
    "r": "Restricted (Rating for mature audiences)",
    "pg": "Parental Guidance (General audiences but parental discretion is advised)",
    "pg-13": "Parents Strongly Cautioned (Some material may be inappropriate for children under 13)",
    "g": "General Audience (Suitable for all audiences)",
    "nc-17": "No One 17 and Under Admitted",
    "vfx": "Visual Effects",
    "fx": "Effects (short for special effects)",
    "cgi": "Computer-Generated Imagery",
    "sfx": "Special Effects",
    "adr": "Automated Dialogue Replacement",
    "dop": "Director of Photography",
    "cin": "Cinematography",
    "op": "Opening (as in opening weekend or opening credits)",
    "bts": "Behind the Scenes",
    "cg": "Computer Graphics",
    "loc": "Location",
    "sound": "Soundtrack",
    "vfx": "Visual Effects",
    "eta": "Estimated Time of Arrival",
    "db": "Database",
    "dba": "Database Administrator",
    "rt": "Runtime",
    "omdb": "Open Movie Database",
    "wip": "Work In Progress",
    "rip": "Rest In Peace",
    "awol": "Absent Without Leave",
    "bom": "Box Office Mojo",
    "us": "United States",
    "uk": "United Kingdom",
    "can": "Canada",
    "fr": "France",
    "de": "Germany",
    "it": "Italy",
    "es": "Spain",
    "in": "India",
    "br": "Brazil",
    "jp": "Japan",
    "kr": "South Korea",
    "usc": "United States of Cinema",
    "fa": "Film Academy",
    "imax": "Image Maximum",
    "vod": "Video on Demand",
    "ott": "Over-The-Top (Content delivery over the internet)"
}
chat_words
```

```python
### FILL Abbreviations

def fill_abbreviations(sentence):
    new_text = []
    for word in sentence.split():
        if word in chat_words:
            new_text.append(chat_words[word])
        else:
            new_text.append(word)
    return " ".join(new_text)

df['overview'] = df['overview'].apply(fill_abbreviations)
```

```python

# REMOVE EMOJIS
import re

def remove_emoji(text):
    emoji_pattern = re.compile("["
                              u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1E0-\U0001F1FF"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                              "]+",flags=re.UNICODE)
    return emoji_pattern.sub(r'',text)

df['overview'] = df['overview'].apply(remove_emoji)
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def stem_words(text):
    stemmed_word_list = []
    for word in text.split():
        stemmed_word = stemmer.stem(word)
        stemmed_word_list.append(stemmed_word)
    return " ".join(stemmed_word_list)


df['overview'] = df['overview'].apply(stem_words)
```
