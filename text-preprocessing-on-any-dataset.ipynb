```python
# TRANSFORM REVIEW COLUMN TO LOWERCASE
df['review'] = df['review'].str.lower()
```

```python
# REMOVING HTML TAGS
import re 

def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'',text)


df['review'] = df['review'].apply(remove_html_tags)
df.head()
```

```python
# REMOVING PUNCTUATIONS IF IT EXISTS IN THE DATA SUCH AS !,#?. etc.

# METHOD 01 (It takes long time duration for removal)
import string, time 
exclude = string.punctuation
exclude

def remove_punctuation1(text):
   for char in exclude:
        text = text.replace(char,' ')
   print(text)

# Dummy Output
start = time.time()
remove_punctuation1("How are you my boy!. It's your house.")
end = time.time()- start
print(end)
        
```

```python
# METHOD 02 (It takes less time duration for removal)

def remove_punctuation2(text):
    return text.translate(str.maketrans('','',exclude))

# Dummy Output
start = time.time()
remove_punctuation2("How are you my boy!. It's your house.")
end = time.time()- start
print(end)
```

```python
# WE USE METHOD 02 FOR OUR PROBLEM AS SOLUTION BECAUSE IT IS OTPIMAL AND CONCISE CODE AND TAKE LESS TIME

df['review'] = df['review'].apply(remove_punctuation2)
```

```python
from nltk.corpus import stopwords
stopword = stopwords.words("english")

def remove_stopwords(text):
    new_text = []
    for word in text:
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return "".join(x)

print(remove_stopwords("Hello, I am Karan Sharma. My elder brother is very nice guy"))
            
```

```python
## IF RIVIEWS CONTAINS ANY TYPE OF EMOJIS TO SHOW HAPPINESS, SORROW, ANGER AND CONFUSION, THERE ARE 2 WAYS TO HANDLE IT
### 1. Remove Emojis
### 2. Replace Emojis To Their Meanings
```

```python
# REMOVE EMOJIS
import re

def remove_emoji(text):
    emoji_pattern = re.compile("["
                              u"\U0001F600-\U0001F64F"
                               u"\U0001F300-\U0001F5FF"
                               u"\U0001F680-\U0001F6FF"
                               u"\U0001F1E0-\U0001F1FF"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                              "]+",flags=re.UNICODE)
    return emoji_pattern.sub(r'',text)

print(remove_emoji("I love my India, ❤️❤️❤️, because it contains many diversifications ⛰️in terms of geogrpahical, religions, languages."))
```

```python
# REPLACE EMOJIS TO THEIR MEANINGS
import emoji

print(emoji.demojize("I love my India, ❤️❤️❤️, because it contains many diversifications ⛰️in terms of geogrpahical, religions, languages.").replace(":",' '))
```

```python
### TOKENIZATIONS: 

## It refers to process of convert sentences or words into tokens (small units)
### Examples: " Humans are greatest creatures of God." --------> ['Humans','are','greates','creatures','of','God']

# Challenges:
# Prefix -> $20
# Suffix -> 5km
# Infix -> Ph.D , abc@gmail.com
# Some Special Cases need speical code to tokenize

# Ways to perform tokenize:
# 1. Split() using python
# 2. Regular expression
# 3. NLTK library
# 4. Spacy library
```

```python
# SPLIT (TOKENIZATION) <===================== example
sent1 ="I am Ph.D holder"
sent2 = "This is my id abc@gmail.com"
sent3 = "My house is 5km far away to current location"
sent4 = "I need $5 for burger"
print(sent1.split())
print(sent2.split())
print(sent3.split())
print(sent4.split())
```

```python
# REGULAR EXPRESSION (TOKENIZATION) <=============== example

import re


# Tokenization using regular expression
tokens1 = re.findall(r"\b\w+\b", sent1)
tokens2 = re.findall(r"\b\w+\b", sent2)
tokens3 = re.findall(r"\b\w+\b", sent3)
tokens4 = re.findall(r"\b\w+\b", sent4)
# Print the tokens
print(tokens1)
print(tokens2)
print(tokens3)
print(tokens4)

```

```python
## NLTK (TOKENIZATION) <================= example

from nltk.tokenize import word_tokenize

print(word_tokenize(sent1))
print(word_tokenize(sent2))
print(word_tokenize(sent3))
print(word_tokenize(sent4))
```

```python
## SPACY (TOKENIZATION)  <====================== example

import spacy

nlp = spacy.load('en_core_web_sm')

def use_spacy(sentence):
    doc = nlp(sentence)
    tokenize_sent = []
    for word in doc:
        tokenize_sent.append(word)
    return tokenize_sent
        
print(use_spacy(sent1))
print(use_spacy(sent2))
print(use_spacy(sent3))
print(use_spacy(sent4))
```

```python
df['review'] = df['review'].apply(use_spacy)
```

```python
# INFLECTION : IT IS PROCESS TO CHANGE A WORD IN SUCH MANNER IT REPRESENT INFORMATION LIKE, TENSE,GENDER,ADJECTIVE,NOUSE ETC. (WALK, WALKED, WALKEN)
# STEMMING : CONVERT A INFLECTED WORD TO ROOT WORD

from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()

def stem_words(text):
    stemmed_word_list = []
    for word in text.split():
        stemmed_word = stemmer.stem(word)
        stemmed_word_list.append(stemmed_word)
    return " ".join(stemmed_word_list)

print(stem_words("walk walked walking walks"))
```
